Importando as bibliotecas e o dataset novamente:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

livros = pd.read_csv('./Dados/GoodReads_100k_books.csv', encoding='utf-8')
```

Dessa vez, para analisar a relação entre a popularidade dos livros e o quão bem eles são avaliados, é possível utilizar um gráfico de dispersão colocando a média das avaliações (ou seja, a nota média que o livro recebe de 0 a 5) contra o total de avaliações publicadas. Nesse caso, podemos utilizar a quantidade de avalições publicadas como uma maneira de mensurar o quão popular dado livro é: quanto mais avaliações, mais leitores que consumiram a obra e se engajaram o suficiente para publicar uma avaliação (seja ela boa ou ruim).

Assim, utilizando um gráfico de dispersão de `totalrating` x `rating`:

```python
sns.FacetGrid(livros, height=7).map(plt.scatter, "totalratings", "rating")
plt.title ("Nota média pela quantidade de avaliações", fontsize=14)
plt.xlabel ("Total de avaliações")
plt.ylabel ("Média das avaliações")
```

Pode-se perceber que, apesar de nem todos os livros bem avaliados se destacarem na quantidade de avaliações, os livros mais populares tendem a possuir notas razoavelmente altas, variando entre aproximadamente 3,5 e 4,5. Nesse sentido, parece existir uma relação entre a popularidade e a nota do livro: livros mais populares tendem a ser bem avaliados, mas não são os melhor avaliados. Possivelmente isso indica que, com mais pessoas avaliando o item, mais notas baixas são atribuídas a ele, mesmo que ele receba, em geral, ótimas avaliações (que provavelmente são responsáveis por sua populari. Assim, a média fica mais "equilibrada".
