```python
df = df.sort_values(by=["rating"],ascending=False)

lista_colunas =['title', 'rating',  'totalratings']



df.dropna(subset = ["title"], inplace=True)

df.head()
```

```python
df=df[~(df['totalratings'] <= 40)]
df1=df.head(50)
```

```python
df1 = Counter(" ".join(df1["title"].str.replace(',', ' ')).split()).most_common(100)
```

```python
df2 = pd.DataFrame(df1, columns= ['Titulo', 'Frequencia'])

df2 = df2.drop(df2.index[(df2["Titulo"] == "and")],axis=0,inplace=False)
df2 = df2.drop(df2.index[(df2["Titulo"] == "I")],axis=0,inplace=False)
df2 = df2.drop(df2.index[(df2["Titulo"] == "Short")],axis=0,inplace=False)
df2 = df2.drop(df2.index[(df2["Titulo"] == "Stories")],axis=0,inplace=False)
df2 = df2.drop(df2.index[(df2["Titulo"] == "Lds")],axis=0,inplace=False)
df2 = df2.drop(df2.index[(df2["Titulo"] == "M")],axis=0,inplace=False)
df2 = df2.drop(df2.index[(df2["Titulo"] == "Self")],axis=0,inplace=False)
df2 = df2.drop(df2.index[(df2["Titulo"] == "Help")],axis=0,inplace=False)
df2 = df2.drop(df2.index[(df2["Titulo"] == "Non")],axis=0,inplace=False)

linha_nova = {"Titulo":"Short Stories","Frequencia":3}
linha_nova2 = {"Titulo":"Self Help","Frequencia":1}

df2 = df2.append(linha_nova, ignore_index=True)
df2 = df2.append(linha_nova2, ignore_index=True)

df2 = df2.sort_values(by=["Frequencia"],ascending=False)
```

```python
df2.plot(x ='Titulo', y='Frequencia', kind = 'bar')
plt.rcParams['figure.dpi'] = 300
plt.rcParams['figure.figsize'] = [30, 15]
```
